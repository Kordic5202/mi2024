```python
import json
import random
from jinja2 import Template
import nbformat as nbf

# Функція для генерації JSON даних на основі заданих меж

def generate_json_data(num_tickets=20):
    fields = [
        "втрат противника у особовому складі",
        "використаних боєприпасів",
        "втрат противника у обладнанні",
        "втрат противника у танках",
        "втрат противника у бронетехніці",
        "втрат противника у артилерії",
        "втрат противника у вантажівках"
    ]
    colors = ["Мундир", "Грозове небо", "Сталь", "Степ", "Марун", "Олива"]
    
        
    data = []
    
    for i in range(1, num_tickets + 1):
        item = {
            "number": i,
            "field1": random.choice(fields),
            "count": random.randint(5, 30),
            "name": f"exam_{random.randint(5, 300)}",
            "Color": random.choice(colors)
        }
        data.append(item)
    
    return json.dumps(data, indent=4, ensure_ascii=False)

# Генерація JSON
json_data = generate_json_data()
data = json.loads(json_data)

# Шаблон Markdown
markdown_template = """
## Завдання на практичну частину екзамену з дисципліни МІАЗ.

### Білет {{ number }}

Створити прототип інформаційно-аналітичної системи для відображення інформації щодо {{ field1 }} за останні {{ count }} днів на основі наданого [шаблону](https://github.com/RomanMykolaichuk/mi2024_exam.git).

1. База даних {{ name }}. Створити таблицю та заповнити її даними за допомогою sql-файлів з шаблону.
2. Backend на основі Python FastAPI (налаштувати параметри підключення до бази даних, та прописати оператор SELECT для вибірки необхідних полів: 
   date, location, [поле відповідно до завдання] as data_field)
3. Frontend на основі:
   - HTML()
   - CSS ()
   - JavaScript ()

4. Кольорова схема має відповідати основному кольору {{ Color }} відповідно до [брендбуку ЗСУ](https://warbooks.com.ua/contents/brendbuk-зсу-керівництво-із-використання-графічного-стилю-у-візуальній-комунікації.pdf).
"""

template = Template(markdown_template)

def generate_tickets(data):
    tickets = []
    for item in data:
        ticket_md = template.render(
            number=item["number"],
            field1=item["field1"],
            count=item["count"],
            name=item["name"],
            Color=item["Color"]
        )
        tickets.append(ticket_md)
    return tickets

# Генеруємо білети
tickets = generate_tickets(data)

# Створення ipynb файлу
nb = nbf.v4.new_notebook()
nb.cells.append(nbf.v4.new_markdown_cell("# Генеровані білети"))
for ticket in tickets:
    nb.cells.append(nbf.v4.new_markdown_cell(ticket))

with open("exam_tickets.ipynb", "w", encoding="utf-8") as f:
    nbf.write(nb, f)

# Вивід першого білета для перевірки
print(tickets[0])
```
